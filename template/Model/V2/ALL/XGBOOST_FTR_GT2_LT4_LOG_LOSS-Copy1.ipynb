```python
# Applying Grid Search to find the best hyper-parameters for our Model
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics.classification import log_loss
from sklearn.metrics import make_scorer
import time

start = time.time()

parameters = {
    'learning_rate': [0.01],
    'n_estimators': np.arange(100, 1000, 30).tolist(),
    'max_depth': np.arange(3, 10).tolist(),
    'min_child_weight': np.arange(1, 10).tolist(),
    'gamma': np.arange(0.01,1,0.03).tolist(),
    'subsample': np.arange(0.5,1,0.05).tolist(),
    'colsample_bytree': np.arange(0.5,1,0.05).tolist(),
    'objective': ['multi:softprob'],
    'scale_pos_weight': [1],
    'reg_alpha':np.arange(0,1,0.03).tolist()
}

grid_search = RandomizedSearchCV(estimator=classifier, 
                           param_distributions=parameters,
                           scoring=make_scorer(log_loss, greater_is_better=False, needs_proba=True),
                           cv=8,
                           n_jobs=-1,
                           verbose=1,
                                 n_iter=10)

grid_search.fit(X_train, y_train)

time.time() - start
```

```python
# Extract best score calculated with the GridSearchCV
best_score = grid_search.best_score_
display(best_score)
```

```python
# Extract best hyper-parameter calculated with the GridSearchCV
best_params = grid_search.best_params_
best_params
```

```python
# Get all results of Grid Search
cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results.to_csv('./tuning/'+model_name+'-'+target+'_'+start_date+'.csv')
cv_results
```

## Evaluating Model

```python
# Train a new classifier using the best parameters found by the grid search
clf = XGBClassifier(
    nthread=4, 
    seed=15,
    learning_rate=best_params['learning_rate'],
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    min_child_weight=best_params['min_child_weight'],
    gamma=best_params['gamma'],
    subsample=best_params['subsample'],
    colsample_bytree=best_params['colsample_bytree'],
    objective=best_params['objective'],
    scale_pos_weight=best_params['scale_pos_weight'],
    reg_alpha=best_params['reg_alpha']
)
clf.fit(X_train, y_train, eval_metric='logloss')
```

```python
# Predict target values
y_pred = clf.predict(X_test)
```

```python
# Predict probabilities
y_probs = clf.predict_proba(X_test)
```
